```python
import torch
import matplotlib.pyplot as plt
from PIL import Image

IMG_PATH = '/home/david/Downloads/wetransfer_imatges_2023-05-12_1337/reduced/001.png'
MAT_PATH = '/home/david/Downloads/wetransfer_imatges_2023-05-12_1337/Color_naming/ColorNaming/w2c.mat'
THRESHOLD = 0.1

COLOR_ORDER = ['Black', 'Blue', 'Brown', 'Grey', 'Green', 'Orange', 'Pink', 'Purple', 'Red', 'White', 'Yellow']
COLOR_CATEGORIES = [[2, 5, 10], [0, 3, 9], [6, 7], [8], [4], [1]]

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
CHECKPOINT_PATH = "./pretrained_models/adobe_dpe/curl_validpsnr_23.073045286204017_validloss_0.0701291635632515_testpsnr_23.584083321292365_testloss_0.061363041400909424_epoch_510_model.pt"

_ = plt.imshow(Image.open(IMG_PATH))
_ = plt.title("Input image")
```

## 1.Color mapping as input

```python
# Utils
from functools import reduce

def analogic_or(masks):
    return reduce(lambda x, y: x | y, masks)

def evaluate(img, convert_uint = False):
    """
    Evaluate the modle per image instance. Image of Batch size 1.
    """
    # Load image and convert to tensor
    if isinstance(img, str):
        img = Image.open(img)
        if img.mode != 'RGB':
            img = img.convert('RGB')

    img = TF.to_tensor(img).to(DEVICE)

    with torch.no_grad():
        img = img.unsqueeze(0)
        img = torch.clamp(img, 0, 1)

        net_output_img_example, _ = net(img)

        net_output_img_example_numpy = net_output_img_example.squeeze(0).data.cpu().numpy()
        net_output_img_example_numpy = ImageProcessing.swapimdims_3HW_HW3(net_output_img_example_numpy)
        return (net_output_img_example_numpy * 255).astype('uint8') if convert_uint else net_output_img_example_numpy
```

```python
import numpy as np
from scipy.io import loadmat

# Load image
im = Image.open(IMG_PATH)
im = np.array(im)

# Load matrix
mat = loadmat(MAT_PATH)['w2c']

index_im = 1+np.floor(im[...,0].flatten()/8).astype(int) + 32*np.floor(im[...,1].flatten()/8).astype(int) + 32*32*np.floor(im[...,2].flatten()/8).astype(int)

prob_maps = []
fig = plt.figure(figsize=(20,2))
fig.suptitle('COLOR PROBABILITY MAPS', fontsize=14)
for idx, w2cM in enumerate(mat.T):
    map = w2cM[index_im].reshape(im.shape[:2])
    prob_maps.append(map)
    plt.subplot(1, mat.shape[1], idx+1)
    plt.imshow(map/np.max(map))  # Normalize mask to [0,1] - Display purpuses
    plt.title(COLOR_ORDER[idx])

# Join color categories
color_masks = []
color_probs = []
fig = plt.figure(figsize=(20, 4))
fig.suptitle('COLOR CATEGORIES', fontsize=14)
for idx, category in enumerate(COLOR_CATEGORIES):
    mask = analogic_or([im * np.expand_dims(np.greater_equal(prob_maps[i], THRESHOLD).astype(int), axis=2) for i in category]).astype(int)
    prob = np.sum([prob_maps[i] for i in category], axis=0)
    color_masks.append(mask)
    color_probs.append(prob)
    plt.subplot(2, len(COLOR_CATEGORIES), idx+1)
    plt.title(' '.join([COLOR_ORDER[col] for col in category]))
    plt.imshow(mask)
    plt.subplot(2, len(COLOR_CATEGORIES), idx+1+len(COLOR_CATEGORIES))
    plt.imshow(prob)
```

```python
# USE CURL
import model
from util import ImageProcessing
import torchvision.transforms.functional as TF

net = model.CURLNet()
checkpoint = torch.load(CHECKPOINT_PATH, map_location=DEVICE)
net.load_state_dict(checkpoint['model_state_dict'])
net.eval()
if DEVICE == 'cuda':
    net = net.cuda()

curl_images = []
fig = plt.figure(figsize=(20, 3))
fig.suptitle('IMAGES AFTER CURLNet', fontsize=14)
for idx, mask in enumerate(color_masks):
    out = evaluate(Image.fromarray(mask.astype(np.uint8)).convert("RGB"), convert_uint=True)       
    curl_images.append(out)
    plt.subplot(1, len(color_masks), idx+1)
    plt.title(' '.join([COLOR_ORDER[col] for col in category]))
    plt.imshow(out)
    

# Blending using prob maps
reconstructed_im = np.sum([np.round((m * np.expand_dims(p, axis=2))).astype('uint8') for m, p in zip(curl_images, color_probs)], axis=0)
fig = plt.figure()
_ = plt.imshow(reconstructed_im)
_ = plt.title("Reconstructed image")
```
